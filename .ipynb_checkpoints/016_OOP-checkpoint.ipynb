```python
 # private variables

class Account:
    counter = 0
    def __init__(self,openig_bal=0):
        Account.counter =  Account.counter +1
        self.id = Account.counter
        self.bal = openig_bal
        self.num_transactions = 0
        self.max_transactions = 6       
        
    def deposit(self,amount):
        if amount > 0 and self.num_transactions < self.max_transactions:
            self.bal = self.bal+amount
            self.num_transactions = self.num_transactions+1
    def withdraw(self,amount):
        if amount > 0 and self.bal>= amount  and self.num_transactions < self.max_transactions:
            self.bal = self.bal - amount
            self.num_transactions = self.num_transactions+1
    def __str__(self):
        return "Ac Number : {} and Bal is Rs. : {} and No of TX : {}".format(self.id,self.bal,self.num_transactions)
class SavingsAccount(Account):
    pass
class CurrentAccount(Account):
    def __init__(self):
        super().__init__(self)
        self.max_transactions = 300
sa1 = SavingsAccount()
ca1 =  CurrentAccount()

sa1.deposit(100)

print(sa1)

sa1.bal = 500

print(sa1)

sa1.deposit(100)
print(sa1)

# Types
# Public : self.bal
# Protected : self._bal warning the developer as not to use this var outside , or change its value directly
# Private : self.__bal

class Account:
    counter = 0
    def __init__(self,openig_bal=0):
        Account.counter =  Account.counter +1
        self.id = Account.counter
        self.__bal = openig_bal #private variable
        self.num_transactions = 0
        self.max_transactions = 6
        
        
    def deposit(self,amount):
        if amount > 0 and self.num_transactions < self.max_transactions:
            self.__bal = self.__bal+amount
            self.num_transactions = self.num_transactions+1
    def withdraw(self,amount):
        if amount > 0 and self.__bal>= amount  and self.num_transactions < self.max_transactions:
            self.__bal = self.__bal - amount
            self.num_transactions = self.num_transactions+1
    def __str__(self):
        return "Ac Number : {} and Bal is Rs. : {} and No of TX : {}".format(self.id,self.__bal,self.num_transactions)
class SavingsAccount(Account):
    pass
class CurrentAccount(Account):
    def __init__(self):
        super().__init__(self)
        self.max_transactions = 300
sa1 = SavingsAccount()
ca1 =  CurrentAccount()

sa1.deposit(100)

sa1.__bal = 500

print(sa1)

print(sa1.__bal)

# NAME MANGLING

# after name mangling 
# sa1.bal  > sa1.__bal > sa1._Account__bal

sa1._Account__bal = 500

print(sa1)

sa1.withdraw(1200)

print(sa1)

# Encapsulation

class PowerPlant:
    def __init__(self,location,capacity,employees):
        self.location = location
        self.capacity = capacity
        self.employees = employees
        


p1 = PowerPlant("Nashik",'700MW',120)

print(p1.location)

p1.location="Thane"
print(p1.location)

class PowerPlant:
    def __init__(self,location,capacity,employees):
        self.__location = location
        self.capacity = capacity
        self.employees = employees
    def __str__(self):
        return "Location is {} and cap is {} with {} employees".format(self.__location,self.capacity,self.employees)
        


p1 = PowerPlant("Nashik",'700MW',120)

p1.__location = "Nagpur"

print(p1.__location)

print(p1)

p1._PowerPlant__location = "Tarapur"

print(p1)
```

```python
class PowerPlant:
    def __init__(self,location,capacity,employees):
        self.__location = location
        self.capacity = capacity
        self.employees = employees
    def __str__(self):
        return "Location is {} and cap is {} with {} employees".format(self.__location,self.capacity,self.employees)
      #getter
    def locate(self):
        print("the location has been accessed,call 10888 for Emp ID who has accessed")
        return self.__location
    #setter
    def locate_change(self,new_location):
        print("the location has been changed,call 10888 if not changed")
        self.__location = new_location

p1 = PowerPlant("Nashik",'700MW',120)
```

```python
# adding withdraw functionality
# adding checks on the fun as not to have neg or overdraft

class Account():
    counter = 0
    def __init__(self,opening_bal=0):
        Account.counter =Account.counter+1
        self.id = Account.counter
        self.bal =opening_bal
    
    def __str__(self):
        return f"Acc no is {self.id} and bal is {self.bal}"
    
    def deposit(self,amount):
        if amount > 0:
            self.bal = self.bal+amount
        
    def withdraw(self,amount):
        if amount > 0 and self.bal >= amount:
            self.bal = self.bal-amount
            return True
        else:
            return False

a1 = Account(100)
a2 = Account()

a1.deposit(50) #bal of a1 will get updated
a1.withdraw(30)
print(a1)

a2.withdraw(30)# neg balance 
print(a2)
```

```python
# using super function to add more attributes in the parent __init__

class Account():
    counter = 0
    def __init__(self,opening_bal=0):
        Account.counter =Account.counter+1
        self.id = Account.counter
        self.bal =opening_bal
        self.num_tx = 0
        self.max_tx = 5
        
    
    def __str__(self):
        return f"Acc no is {self.id} and bal is {self.bal}"
    
    def deposit(self,amount):
        if amount > 0 and self.num_tx < self.max_tx:
            self.bal = self.bal+amount
            self.num_tx = self.num_tx+1
        
    def withdraw(self,amount):
        if amount > 0 and self.bal >= amount and self.num_tx < self.max_tx:
            self.bal = self.bal-amount
            self.num_tx = self.num_tx+1
            return True
        else:
            return False

# creating two classese and inheriting the Account class
class SavingsAccount(Account):
    pass
class CurrentAccount(Account):
    def __init__(self):
        super().__init__()
        self.max_tx = 100

        
c1= CurrentAccount()
c1.deposit(100)
print(c1)
```
