```python
import os

def get_folder_sizes(path):
    folder_sizes = {}
    for root, dirs, files in os.walk(path):
        size = sum(os.path.getsize(os.path.join(root, name)) for name in files)
        folder_sizes[root.replace('/', '\\')] = size
    return folder_sizes
folder_sizes = get_folder_sizes(r'C:\Users\RAHUL\Documents\OnePython')
for folder, size in folder_sizes.items():
    print(f"{folder}: {size} bytes")
```
