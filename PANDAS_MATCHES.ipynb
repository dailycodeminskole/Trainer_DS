```python
# sorting based on city
data.sort_values('city')
# data.sort_values('city',ascending=False) # for descendnig vlues
```

```python
# sorting by 2 parameters
data.sort_values(['city','date']) #both in ascending by default
data.sort_values(['city','date'],ascending=[True,False]) #city  in ascending and date in descending
```
