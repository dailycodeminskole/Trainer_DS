# creating more complex plots

```python
# plotting a square function
x_val = [-3,-2,-1,0,1,2,3]
y_val = [9,4,1,0,1,4,9]
plt.plot(x_val,y_val)
# no smooth values 
```

# ticker

```python
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.xticks(np.arange(-3,4,.3))

plt.plot(x_val,y_val)
```

```python
# cluttered ticks 
# rotating the angle to 90degrees
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.xticks(np.arange(-3,4,.3),rotation=90)

plt.plot(x_val,y_val)
```

# making plot more RICH

```python
# adding labels and titles
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.title('Sq Function')

plt.xlabel("X_val")
plt.ylabel("Y_val")


plt.xticks(np.arange(-3,4,.3),rotation=90)

plt.plot(x_val,y_val)
plt.show()
```

```python
# changing the color
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.title('Sq Function')

plt.xlabel("X_val")
plt.ylabel("Y_val")


plt.xticks(np.arange(-3,4,.3),rotation=90)

plt.plot(x_val,y_val,color='g')
plt.show()
```

```python
# changing the plain curve using different point markers
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.title('Sq Function')

plt.xlabel("X_val")
plt.ylabel("Y_val")


plt.xticks(np.arange(-3,4,.3),rotation=90)

# plt.plot(x_val,y_val,'g*')#this is not color but green star
plt.plot(x_val,y_val,'ro')#this is not color but red o

plt.show()
```

# zoom

```python
# zoom in to a particulr section of the plot
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.title('Sq Function')

plt.xlabel("X_val")
plt.ylabel("Y_val")


plt.xticks(np.arange(-3,4,.3),rotation=90)

# adding the limits
plt.xlim(2,4)

# adding the limits
plt.ylim(8,14)

# plt.plot(x_val,y_val,'g*')#this is not color but green star
plt.plot(x_val,y_val,'ro')#this is not color but red o

plt.show()
```

```python
# zoom out to a particulr section of the plot
#  birds eye view
x_val = np.arange(-3,4,.1)
y_val = x_val**2

plt.title('Sq Function')

plt.xlabel("X_val")
plt.ylabel("Y_val")


plt.xticks(np.arange(-3,4,.3),rotation=90)

# adding the limits
plt.xlim(-10,10)

# adding the limits
plt.ylim(0,20)

# plt.plot(x_val,y_val,'g*')#this is not color but green star
plt.plot(x_val,y_val,'ro')#this is not color but red o

plt.show()
```

# multiple plots in same canvas

```python
x_val = np.arange(-3,4,.1)
y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
```

# adding legend

```python
x_val = np.arange(-3,4,.1)
y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val,label='sq')
plt.plot(x_val,y2_val,label='cube')
plt.legend() #thhis is to make legend vidsible

plt.show()
```

```python
# alternate way
x_val = np.arange(-3,4,.1)
y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
plt.legend(['sq','cube'])

plt.show()
```

```python
# changing the location of the legend
x_val = np.arange(-3,4,.1)
y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
# plt.legend(['sq','cube'],loc='lower right')
plt.legend(['sq','cube'],loc='lower right',frameon=False) #legends withhout border
plt.legend(['sq','cube'],loc='lower right',frameon=False,ncol=2) #legends in different columns



plt.show()
```

```python
# labelling points on the chart
x_val = np.arange(-3,4,.1)
y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
# plt.legend(['sq','cube'],loc='lower right')
plt.legend(['sq','cube'],loc='lower right',frameon=False) #legends withhout border
plt.legend(['sq','cube'],loc='lower right',frameon=False,ncol=2) #legends in different columns

plt.text(0,0,'one')
plt.text(1,1,'two')

plt.grid()
plt.show()
```

# SCATTER PLOT

```python
x = np.arange(20)
y1 = np.linspace(1,20,20)+np.random.rand(20)*5
y2 = np.linspace(1,20,20)+np.random.rand(20)*10
plt.scatter(x,y1)
plt.scatter(x,y2)
```

```python
# changing The marker size of plots
x = np.arange(20)
y1 = np.linspace(1,20,20)+np.random.rand(20)*5
y2 = np.linspace(1,20,20)+np.random.rand(20)*10
plt.scatter(x,y1,s=y1)
plt.scatter(x,y2,s=y1)
plt.show()
```

#  3D Plots

```python
a = np.array([0,1,2,3])
b = np.array([0,1,2])
a,b = np.meshgrid(a,b)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,np.array([[7]]))
plt.show()
```

# Complex 3D

```python
a = np.arange(-1,1,.005)
b = a
a,b = np.meshgrid(a,b)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,a**2+b**2)
plt.show()
```
