# enriching the graph

```python
x_val = np.arange(-10,10,.5)
y_val = x_val**2
plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y_val,color='c')
plt.show()
```

```python
x_val = np.arange(-10,10,.5)
y_val = x_val**2
plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y_val,'g*')#green with star patter, do not use the key 'color='
plt.show()
```

# multiple plots in the same canvas

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3

plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y1_val) #square func
plt.plot(x_val,y2_val) #cube func

plt.show()
```

# adding legends

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3

plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y1_val,label='Square') #square func
plt.plot(x_val,y2_val,label='Cube') #cube func
plt.legend()

plt.show()
```

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3

plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y1_val) #square func
plt.plot(x_val,y2_val) #cube func
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)

plt.show()
```

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3

plt.xticks(np.arange(-10,10,.5),rotation=90)
plt.plot(x_val,y1_val) #square func
plt.plot(x_val,y2_val) #cube func
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)

plt.text(0,0,"one")
plt.text(1,1,"two")


plt.xlim(-2,2)
plt.ylim(-2,2)
plt.grid() #to make the grid visible 
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load iris dataset
iris = sns.load_dataset('iris')

# Create subsets for each species
setosa = iris.loc[iris['species'] == 'setosa']
versicolor = iris.loc[iris['species'] == 'versicolor']
virginica = iris.loc[iris['species'] == 'virginica']

# Create scatter plot
plt.scatter(x=setosa['petal_length'], y=setosa['petal_width'], c='blue', label='Setosa')
plt.scatter(x=versicolor['petal_length'], y=versicolor['petal_width'], c='orange', label='Versicolor')
plt.scatter(x=virginica['petal_length'], y=virginica['petal_width'], c='green', label='Virginica')

# Add labels and legend
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Dataset')
plt.legend()

# Show plot
plt.show()
```
